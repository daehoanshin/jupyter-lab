```python
import pandas as pd
a = pd.read_csv('data/mtcars.csv', index_col=0)
```

```python
# 사용자 코딩
from sklearn.preprocessing import MinMaxScaler
scaler = ['qsec']
min = MinMaxScaler()
min.fit(a[scaler])
a[scaler] = min.transform(a[scaler])

print(a[scaler])
```

```python
result = a[a['qsec'] > 0.5]
print('결과값은', result)
```
